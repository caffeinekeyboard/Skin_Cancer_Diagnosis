#### Converting of the implementation of InceptionResNetV2 for classification of skin lesions using the MNIST-HAM10000 Dataset into different formats.
- Date : January 30th 2024
- Author : Aarya Bhave
- Project : Skin_Cancer_Diagnosis
  
This code carries out conversion of the model into different file formats.  
Before running this file locally, you must have Tensorflow-GPU installed with a dedicated GPU.
 

```python
import keras as K 
import tensorflow as tf 
```

```python
incres_model = K.models.load_model('InceptionResNetV2_Loadable')
```

```python
incres_model.summary()
```

```python
incres_model.save('Prototype_Models/InceptionResNetV2_H5.h5')
```

```python
h5_model = K.models.load_model('Prototype_Models/InceptionResNetV2_H5.h5')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(h5_model)
tflite_model = converter.convert()
open("Prototype_Models/InceptionResNetV2_Lite.tflite", "wb").write(tflite_model)
```

```python
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model_optimized = converter.convert()
open("Prototype_Models/InceptionResNetV2_Lite_Optimized.tflite", "wb").write(tflite_model)
```
