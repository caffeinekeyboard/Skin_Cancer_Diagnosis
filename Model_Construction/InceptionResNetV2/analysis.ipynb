```python
import seaborn as sns
import matplotlib.pyplot as plt 
import cv2
import pandas as pd
```

##### Meta Data 

```python
meta_data = pd.read_csv('data/HAM10000_metadata.csv')
meta_data
```

##### Diagnosis Distributions

```python
sns.countplot(x=meta_data['dx'], order=meta_data['dx'].value_counts().index)
plt.show()
print(meta_data['dx'].value_counts())
```

##### Diagnosis Type Distributions

```python
sns.countplot(x=meta_data['dx_type'], order=meta_data['dx_type'].value_counts().index)
plt.show()
print(meta_data['dx_type'].value_counts())
```

##### Sex Distributions

```python
sns.countplot(x=meta_data['sex'], order=meta_data['sex'].value_counts().index)
plt.show()
print(meta_data['sex'].value_counts())
```

##### Localization

```python
plot = sns.countplot(x=meta_data['localization'], order=meta_data['localization'].value_counts().index)
plot.set_xticklabels(plot.get_xticklabels(), rotation=40, ha='right')
plt.show()
print(meta_data['localization'].value_counts())
```

##### Age Distributions

```python
plot = sns.relplot(data=meta_data['age'])
plt.show()
print(meta_data['age'].value_counts())
```
